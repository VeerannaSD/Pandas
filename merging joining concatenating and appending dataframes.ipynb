```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,5,9],
                 'col_1':[1,2,3,4,5],
                 'col_2':[6,7,8,9,19],
                 'col_3':[11,12,13,14,15],
                  'col_4':['apple','orange','banana','strawberry','raspberry']
                 })
```

```python
df2=pd.DataFrame({'ID':[1,1,3,5],
                 'col_A':[8,9,10,11],
                 'col_B':[12,13,15,17],
                # 'col_3':[12,13,14,15],
                  'col_4':['apple','orange','banana','kiwi']
                 })
```

```python
df1
```

```python
df2
```

```python
inner=pd.merge(df1,df2)
inner
```

#Merging based on ID

```python
pd.merge(df1,df2,on='ID')
```

#Merging based on two columns names ID and col_4

```python
pd.merge(df1,df2,on=['ID','col_4'])
```

#Suffixes and merging on columns that are unique to each dataframe

```python
pd.merge(df1,df2,suffixes=['_l','_r'],left_on='col_2',right_on='col_A')
```

Merging on Indexes

```python
pd.merge(df1,df2,suffixes=['_l','_r'],left_index=True,right_index=True)
```

#Inner Join pd.merge(df1,df2,on='col_name') above snippets are example of this

#Outer Join pd.merge(df1,df2,on='col_name',how='outer')

```python
pd.merge(df1,df2,on='col_4',how='outer',suffixes=['_l','_r'])
```

#left join  pd.merge(df1,df2,on='col_name',how='left') keep data which matched with columns on right hand dataframe and fill NaN inplace of not matched, keep all left dataframe as is

```python
pd.merge(df1,df2,on='col_4',how='left',suffixes=['_l','_r'])
```

#right join  pd.merge(df1,df2,on='col_name',how='right')
#keep data which matched with columns on left hand dataframe and fill NaN inplace of not matched, keep all right# dataframe as is

```python
pd.merge(df1,df2,on='col_4',how='right',suffixes=['_l','_r'])
```

#Join dataframes

```python
df1.join(df2,on='ID',lsuffix='_l',rsuffix='_r')
```

```python
df1.join(df2,on='ID',how='inner',lsuffix='_l',rsuffix='_r')
```

```python
#Concat dataframes
#Default axis is 0 stacks dataframes 
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)#Indexes reset to normat count
```

```python
#Concat dataframes side by side
```

```python
pd.concat([df1,df2],axis=1)
```

```python
#Default join join is outer but we can specify inner join
#ther is no option for left or right join
```

```python
pd.concat([df1,df2],axis=1,join='inner')
```

```python
pd.concat([df1,df2],axis=1,join='outer')
```

```python
pd.concat([df1,df2],axis=0,join='inner')
```
