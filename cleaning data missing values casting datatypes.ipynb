```python
import pandas as pd
import numpy as np
```

```python
people={'first':['Corey','Jane','John','Chris',np.nan,None,'NA'],
       'last':['Schafer','Doe','Doe','Schafer',np.nan,np.nan,'Missing'],
       'email':['CoreyMSchafer@gmail.com','JaneDoe@email.com','JonDoe@email.com',None,np.nan,'Ananymous@email.com','NA'],
       'age':['33','55','63','36',None,None,'Missing']}
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
df.dropna()#Drop missing values
```

```python
df.dropna(axis='index',how='any')#These arguments are default values used bu dropna axis is row and drop if any value in column as NA
```

```python
#how set to all which means it checks all data on row are NA thats why rwo 4 is dropped where all data are NA
df.dropna(axis='index',how='all')
```

```python
#Dropna based on column, here checks if all values in column are NA
df.dropna(axis='columns',how='all')
```

```python
#Drop NA if any of the column has NA or none,here all columns have NA or none so all columns are dropped
df.dropna(axis='columns',how='any')
```

Replacing null/NA values with meaningful values

```python
df1=pd.DataFrame(people)
#Replacing NA/Missing values with numpy nan values
df1.replace('NA',np.nan,inplace=True)
df1.replace('Missing',np.nan,inplace=True)
df1
```

```python
df1.dropna()
```

```python
df1.isna()#Checks value in dataframe a present or not
```

fillna method

```python
#Fill null values with 0
df1.fillna(0)
```
