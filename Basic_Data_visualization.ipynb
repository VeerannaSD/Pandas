```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv('heart.csv')
df.head()
```

Bar Plots

```python
chest_pain=df.groupby(by='cp').median().reset_index()
chest_pain
```

```python
#Bar plot using python
chest_pain.plot.bar(x='cp',y='age');
```

```python


chest_pain.plot.bar(x='cp');
```

```python
#Bar plot using matplotlib

fig,ax=plt.subplots()
ax.bar(chest_pain['cp'],chest_pain['age'],label='age')
ax.set_xlabel('cp')
ax.set_ylabel('age')
ax.legend();
```

```python
fig,ax=plt.subplots()
ax.bar(chest_pain['cp'],chest_pain['age'],label='age',edgecolor='k',color=['red','blue','green','black'])
ax.set_xlabel('cp')
ax.set_xticks(chest_pain['cp']);
```

```python
ax=sb.barplot('cp','age',errcolor='w',capsize=0.1,data=df)
```

Scatter Plot

```python
df.plot.scatter('age','trtbps');
```

```python
fig,ax=plt.subplots()
ax.scatter(df['age'],df['trtbps'],marker='.',s=60,c=df['age'],edgecolors='k',alpha=.5)
ax.set_xlabel('age')
ax.set_ylabel('trtbps');
```

```python
sb.scatterplot('age','trtbps',hue='age',s=30,edgecolor='none',data=df);
```

Line Plots

```python
ages=df.groupby('age').median().reset_index()
ages.head()
```

```python
ages.plot.line('age',['chol','trtbps'])
```

```python
fig,ax=plt.subplots()
ax.plot(ages['age'],ages['chol'],ls=':',lw=1.7)
ax.set_xlabel('age')
ax.set_ylabel('chol')
```

```python
sb.lineplot('age','chol',data=ages);
```
